# Model Fitting Using a GLM 
The general model will have the form:
$$ Sp_{rich}(x,y) = \beta_1 AET + \beta_2 Age + Z(x,y) + \epsilon $$
Where:
$\beta_1$ and $\beta_2$ are model parameters, $Z(x,y)$ is the *Spatial Autocorrelation* process and $\epsilon \sim N(0,\sigma^2)$

```python

len(data.lon)
#X = data[['AET','StandAge','lon','lat']]
X = data[['SppN','lon','lat']]
X = data[['lon','lat']]
Y = data['plotBiomass']
## Let´s take a small sample only for the spatial autocorrelation
import numpy as np
randindx = np.random.randint(0,X.shape[0],1000)
nX = X.loc[randindx]
nY = Y.loc[randindx]
```

```python
# Import GPFlow
import GPflow as gf
k = gf.kernels.Matern12(2, lengthscales=1, active_dims = [1,2] )
```

```python
model = gf.gpr.GPR(nX.as_matrix(),nY.as_matrix().reshape(1000,1).astype(float),k)
```

```python
model.likelihood.variance = 10
model.optimize()
```

```python
Nn = 300
predicted_x = np.linspace(min(data.lon),max(data.lon),Nn)
predicted_y = np.linspace(min(data.lat),max(data.lat),Nn)
Xx, Yy = np.meshgrid(predicted_x,predicted_y)
## Fake richness
fake_sp_rich = np.ones(len(Xx.ravel()))
predicted_coordinates = np.vstack([fake_sp_rich, Xx.ravel(), Yy.ravel()]).transpose()
```

```python
means,variances = model.predict_y(predicted_coordinates)
```

```python
plt.pcolor(Xx,Yy,means.reshape(Nn,Nn))

```

```python
ndata = pd.concat([nY,nX],axis=1)
plt.scatter(ndata.lon,ndata.lat,c=ndata.SppN)
plt.title("Species Richness")
plt.colorbar()
```
